```python
url = "https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&diff=945633050&oldid=942851379"
df =pd.read_html(url, header=0)[1]
df.head()
```

```python

df = df[ df["Borough"] != "Not assigned"]

df
```

```python
df_fine = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
```
